### 4b. Register images as a custom mini-dataset

DFoT's data loader expects a metadata `.pt` file that lists video paths and
matching camera poses. We create a minimal synthetic version: an 8-frame
'video' where frame 0 and frame 7 are your images and frames 1вЂ“6 are
black placeholders (the model will fill those in). The camera poses follow
a smooth **forward-motion trajectory** вЂ” tune the parameters below.

| Parameter | Effect |
|---|---|
| `Y_ANGLE` | Horizontal pan вЂ” positive = pan right, negative = pan left |
| `X_ANGLE` | Vertical tilt вЂ” positive = tilt up, negative = tilt down |
| `DISTANCE` | Forward movement вЂ” 0 = static, 100 = moderate, 200 = large |

```
# в”Ђв”Ђ Trajectory parameters в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ в†ђ tune these!
Y_ANGLE  =   0.0   # horizontal pan  (degrees, + = right, - = left)
X_ANGLE  =   0.0   # vertical tilt   (degrees, + = up,    - = down)
DISTANCE = 100.0   # forward motion  (0вЂ“200)
N_FRAMES =   8     # total frames in output video (keep at 8 for this task)

# в”Ђв”Ђ Build poses в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
import torch, math, numpy as np
from scipy.spatial.transform import Rotation, Slerp

def build_trajectory(n_frames, x_angle=0.0, y_angle=0.0, distance=100.0):
    """Returns a (n_frames, 16) pose tensor in RE10K format."""
    # Intrinsics: (fx, fy, cx, cy) normalised вЂ” RE10K convention
    K = torch.tensor([0.5, 0.5, 0.5, 0.5])

    MOVING_SCALE = 0.5 * distance / 100.0

    R_start = np.eye(3, dtype=np.float32)
    T_start = np.zeros(3, dtype=np.float32)

    # Final pose rotation (using RE10K euler convention xyz)
    R_final = Rotation.from_euler(
        "xyz", [-x_angle, -y_angle, 0.0], degrees=True
    ).as_matrix().astype(np.float32)

    T_final = np.array([
        -MOVING_SCALE * n_frames * math.sin(math.radians(y_angle)),
         MOVING_SCALE * n_frames * math.sin(math.radians(x_angle)),
        -MOVING_SCALE * n_frames * math.cos(math.radians(y_angle)),
    ], dtype=np.float32)

    ts = np.linspace(0.0, 1.0, n_frames)
    slerp = Slerp([0.0, 1.0], Rotation.from_matrix(np.stack([R_start, R_final])))
    R_interp = slerp(ts).as_matrix().astype(np.float32)   # (n_frames, 3, 3)
    T_interp = np.outer(ts, T_final).astype(np.float32)   # (n_frames, 3)

    RT = np.concatenate([R_interp, T_interp[:, :, None]], axis=-1)  # (n_frames, 3, 4)
    RT_flat = RT.reshape(n_frames, 12)

    K_rep = K.unsqueeze(0).expand(n_frames, -1).numpy()   # (n_frames, 4)
    poses = torch.from_numpy(np.concatenate([K_rep, RT_flat], axis=-1))  # (n_frames, 16)
    return poses


poses = build_trajectory(N_FRAMES, x_angle=X_ANGLE, y_angle=Y_ANGLE, distance=DISTANCE)
print(f"вњ… Trajectory shape: {poses.shape}  (n_frames=8, pose_dim=16)")
print(f"   Pan={Y_ANGLE}В°  Tilt={X_ANGLE}В°  Distance={DISTANCE}")
```

```
# в”Ђв”Ђ Build the synthetic video tensor and save metadata в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
import torchvision.transforms as T

transform = T.ToTensor()   # PIL в†’ (C, H, W) float32 in [0, 1]

f0 = transform(img1)  # your start frame вЂ” (3, 256, 256)
f7 = transform(img2)  # your end   frame вЂ” (3, 256, 256)

# Frames 1-6 are black; DFoT's interpolation task will generate them
black = torch.zeros_like(f0)
video = torch.stack([f0, black, black, black, black, black, black, f7])  # (8, 3, 256, 256)

# Save video and poses for the data loader
CUSTOM_VIDEO_PT  = CUSTOM_DIR / "video.pt"
CUSTOM_POSES_PT  = CUSTOM_DIR / "poses.pt"
CUSTOM_META_PT   = CUSTOM_DIR / "metadata.pt"

torch.save(video,  CUSTOM_VIDEO_PT)
torch.save(poses,  CUSTOM_POSES_PT)

# Metadata dict expected by the RE10K mini loader
metadata = {
    "video_paths":  [CUSTOM_VIDEO_PT],
    "pose_paths":   [CUSTOM_POSES_PT],
}
torch.save(metadata, CUSTOM_META_PT)

print("вњ… Saved:")
print(f"   video   в†’ {CUSTOM_VIDEO_PT}")
print(f"   poses   в†’ {CUSTOM_POSES_PT}")
print(f"   metadataв†’ {CUSTOM_META_PT}")
```

## Step 5 вЂ” Run DFoT interpolation

We run the **two-image interpolation** task using the official CLI.
The command is identical to the working baseline, except:
- `dataset=realestate10k_mini` is kept (the pretrained model expects this config)
- `dataset.num_eval_videos=1` вЂ” only our one custom scene
- `dataset.frame_skip=1` + `dataset.n_frames=8` вЂ” 8 output frames
- `WANDB_MODE=offline` so no network is needed during generation

> вЏ± Expect **~30вЂ“60 seconds** on an A100.

```
# Patch the realestate10k_mini dataset config to point at our custom data
import re, pathlib

ds_cfg_path = pathlib.Path("configurations/dataset/realestate10k_mini.yaml")
ds_text = ds_cfg_path.read_text()
print("--- Original dataset config ---")
print(ds_text)
print("-------------------------------")
```

```
# Override the data root in the dataset config to point at our custom folder
# We write a tiny override yaml that Hydra will merge on top
import pathlib, textwrap

override_path = pathlib.Path("configurations/dataset/custom_override.yaml")
override_path.write_text(textwrap.dedent(f"""
    # Auto-generated override вЂ” points to custom images
    data_dir: data/custom
    metadata_dir: data/custom
""").strip() + "\n")
print(f"Wrote override: {override_path}")
print(override_path.read_text())
```

```
# в”Ђв”Ђ Main generation command в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
# Uses the same flags as the working baseline notebook, plus:
#   dataset.num_eval_videos=1   в†’ only evaluate our one custom scene
#   dataset.frame_skip=1        в†’ consecutive frames (we only have 8)
# WANDB_MODE=offline so the run logs locally without needing a network call

!WANDB_MODE=offline python -m main \
    +name=custom_two_images \
    dataset=realestate10k_mini \
    algorithm=dfot_video_pose \
    experiment=video_generation \
    @diffusion/continuous \
    load=pretrained:DFoT_RE10K.ckpt \
    'experiment.tasks=[validation]' \
    experiment.validation.data.shuffle=False \
    dataset.frame_skip=1 \
    dataset.n_frames=8 \
    experiment.validation.batch_size=1 \
    dataset.num_eval_videos=1 \
    algorithm.tasks.prediction.enabled=False \
    algorithm.tasks.interpolation.enabled=True \
    algorithm.tasks.interpolation.history_guidance.name=vanilla \
    +algorithm.tasks.interpolation.history_guidance.guidance_scale=4.0 \
    'algorithm.logging.metrics=[]'
```
